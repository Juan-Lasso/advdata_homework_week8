```python
##Import our library 

import pandas as pd 
```

```python
## read and store csv in a dataframe
mfs = pd.read_csv("raw_data/motor_accidents_raw.csv", low_memory = False)
```

```python
## see the overall info about this dataset
mfs.info()
```

```python
## create a series for borough using dot notation
borough_series = mfs.BOROUGH
borough_series
```

```python
## create a series for borough using dot notation
crash_date_series = mfs["CRASH DATE"]
crash_date_series
```

```python
## create a subset of data for only Queens
## place it in a dataframe called df_q
df_q = mfs.query("BOROUGH == 'QUEENS'")
```

```python
df_q
```

```python
## create a dataset for Manhattan that involved taxi cabs as the primary vehicle cause
df_m = mfs.query("BOROUGH == 'MANHATTAN' and `VEHICLE TYPE CODE 1` == 'Taxi'")
```

```python
df_m
```

```python
## find all incidents of defective pavements causing accidents 
## and sort by borough
query_values = "(`CONTRIBUTING FACTOR VEHICLE 1` == 'Pavement Defective') | \
             (`CONTRIBUTING FACTOR VEHICLE 2` == 'Pavement Defective') | \
             (`CONTRIBUTING FACTOR VEHICLE 3` == 'Pavement Defective') | \
             (`CONTRIBUTING FACTOR VEHICLE 4` == 'Pavement Defective') | \
             (`CONTRIBUTING FACTOR VEHICLE 5` == 'Pavement Defective')"
dfp = mfs.query(query_values)
dfp_sorted = dfp.sort_values(by='BOROUGH')
dfp_sorted
```

```python
## find all incidents in which more than 3 people were killed
df_over_3 = mfs.query("`NUMBER OF PERSONS KILLED` > 3")
df_over_3
```

```python
## find all incidents in which between 2 and 3 people were killed
df_btwn = mfs.query("2 <= `NUMBER OF PERSONS KILLED` <= 3")
df_btwn
```
